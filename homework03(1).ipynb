### Problem 3.1 (Exam Style)

In the case of $1$ dimension, the aperture problem can occur when the intensity profile is locally flat (so $\frac{\partial I}{\partial x} = 0$, which causes the BCCE to be unable to give any information about the motion $u$ i.e. any velocity $u$ satisfies $\frac{\partial I}{\partial x}u + \frac{\partial I}{\partial t} = 0$ because $\frac{\partial I}{\partial x} = 0$. In terms of the images, it means that a purely uniform or constant‐brightness region cannot show how fast it is moving, because shifting it left or right produces the same measured brightness.

### Problem 3.2

```python
def central_differences(v):
    n = len(v)
    d = np.zeros_like(v, dtype=float)
    for x in range(1, n-1):
        d[x] = (v[x+1] - v[x-1]) / 2
    d[0]   = d[1]
    d[n-1] = d[n-2]
    return d

spat_deriv = central_differences(frame0)
time_deriv = frame1 - frame0

u = np.full_like(frame0, np.nan, dtype=float)
nonzero_mask = (spat_deriv != 0)
u[nonzero_mask] = - time_deriv[nonzero_mask] / spat_deriv[nonzero_mask]

plt.figure(figsize=(10,6))
plt.subplot(2, 1, 1)
plt.plot(spat_deriv, label=r'$\partial I/\partial x$')
plt.plot(time_deriv, label=r'$\partial I/\partial t$')
plt.xlabel('pixel')
plt.ylabel(r'$\partial I/\partial x$ or $\partial I/\partial t$')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(u, label='u')
plt.xlabel('pixel')
plt.ylabel(r'$\partial I/\partial x$ or $\partial I/\partial t$')
plt.ylim(-1, 1)
plt.legend()

plt.show()
```

```python
def finite_differences(v):
    d = np.diff(v)
    return np.append(d, d[-1])
```

```python
from autograd import numpy as anp
from autograd import grad
from scipy.optimize import minimize
```

```python
class Cost:
    def __init__(self, f, g, mu):
        self.f = f
        self.g = g
        self.mu = mu
        df = finite_differences(f)
        dg = finite_differences(g)
        self.Ix = (df + dg) / 2.0
        self.It = g - f          

    def cost(self, u):
        data_residual = self.Ix * u + self.It  # element-wise
        data_cost = anp.sum(data_residual ** 2) / len(u)
        du = finite_differences(u)
        smooth_cost = anp.sum(du ** 2) / len(u)
        return data_cost + mu * smooth_cost

mu = 1.e3
cost = Cost(frame0, frame1, mu)
u_init = anp.zeros_like(frame0, dtype=float)
u_star = minimize(cost.cost, u_init, jac=grad(cost.cost),
options={'gtol': 1e-4, 'maxiter': 1000}).x
columns = [31, 72, 118]
for i in columns:
    print("Column ", i, ": ", f"{u_star[i]:.3f}" "\n")

def plot_horn_schunk_results(epi, k0, u_star, columns):
    width = epi.shape[1]
    x_coords = np.arange(width)
    
    start_row = max(0, k0 - 20)
    end_row   = min(epi.shape[0]-1, k0 + 20)
    epi_sub = epi[start_row:end_row+1, :]

    plt.figure(figsize=(10,6))
    plt.subplot(2, 1, 1)
    plt.imshow(epi_sub, cmap='gray')
    plt.xlabel("X (pixel)")
    plt.ylabel("Row (frame index)")
    plt.yticks([0, 20, 40], ["280", "300", "320"])

    plt.axhline(y=k0 - start_row, color='red', label=f"Frame {k0}")

    for c in columns:
        plt.plot(c, k0 - start_row, 'ro')

    plt.xlim(0, width)
    
    plt.subplot(2, 1, 2)
    plt.plot(x_coords, u_star, label=r"$u^*$")
    newtext = f"Horn–Schunk Flow, $u^*$, with regularization parameter, {mu}"
    plt.title(newtext)
    plt.xlabel("X (pixel)")
    plt.ylabel(r"Flow magnitude, $u^*$")

    for c in columns:
        plt.plot(c, u_star[c], 'ro')

    plt.xlim(0, width)
    plt.show()

plot_horn_schunk_results(epi, k0, u_star, columns)
```

$\delta k$ measures how many frames elapse between two images, while $x$ measures how far (in pixels) a feature has shifted horizontally.  If we assume (locally) a constant velocity $v$ in pixels per frame, then after $\delta k$ frames the feature has shifted by
$$
\Delta x \;=\; v\,\delta k.
$$
If $x=0$ is the feature’s initial position, then we have
$$
x(\delta k)
\;=\;
x(0) \;+\; v\,\delta k
\;\approx\;
v\,\delta k.
$$
The image‐space coordinate $x$ is proportional to the frame difference $\delta k$, with proportionality constant $v$.  Equivalently, 
$$
\frac{x(\delta k)}{\delta k}
\;\approx\;
v,
$$
which is just the (approximate) horizontal velocity in pixels/frame.  It is a linear relationship.

```python
# Task 1

def gaussian_1d_window(h_win_size, sigma=None):
    if sigma is None:
        sigma = h_win_size / 3.0
    xs = np.arange(-h_win_size, h_win_size + 1, dtype=float)
    w = np.exp(-0.5 * (xs / sigma)**2)
    return w / w.sum()

class LucasKanade1d:
    def __init__(self, f, g, h_win_size=6, min_step=1.e-4, max_iter=1000, small=1.e-4):
        self.f = f.astype(float)
        self.g = g.astype(float)
        self.n = len(f)
        self.h_win  = h_win_size
        self.min_step = min_step
        self.max_iter = max_iter
        self.small = small
        self.window = gaussian_1d_window(h_win_size)

    def track(self, p):
        if p < self.h_win or p > self.n - 1 - self.h_win:
            return None

        q = float(p)

        for _iter in range(self.max_iter):
            center = int(round(p))
            coords_int = np.arange(center - self.h_win, center + self.h_win + 1)
            if coords_int[0] < 0 or coords_int[-1] >= self.n:
                return None
            win_f = self.f[coords_int]
            shift = q - p
            g_coords = coords_int + shift
            if g_coords[0] < 0 or g_coords[-1] > self.n - 1:
                return None
            win_g = np.interp(g_coords, np.arange(self.n), self.g)
            g_deriv = central_differences(self.g)
            win_gx  = np.interp(g_coords, np.arange(self.n), g_deriv)

            w = self.window

            a = np.sum(w * (win_gx ** 2))
            b = np.sum(w * win_gx * (win_f - win_g))

            if a < self.small:
                return None

            d = b / a
            new_q = q + d

            if (new_q < self.h_win or new_q > self.n - 1 - self.h_win
                or abs(new_q - p) > self.h_win):
                return None

            if abs(d) < self.min_step:
                return new_q

            q = new_q
        return q

#----------------------------------------------------------------------------------------------------
# Task 2

k0 = 300
k_end = 349
columns = [31, 72, 118]

num_points = 3
num_frames = (k_end - k0 + 1)

trajectories = np.zeros((num_points, num_frames), dtype=float)

for i in range(num_points):
    trajectories[i, 0] = columns[i]

for k in range(k0, k_end):
    f = epi[k, :]
    g = epi[k + 1, :]
    for i in range(num_points):
        p = trajectories[i, k - k0]
        lk = LucasKanade1d(f, g)
        q = lk.track(p)
        if q is None:
            trajectories[i, (k+1 - k0):] = np.nan
            break
        trajectories[i, k+1 - k0] = q
        
#----------------------------------------------------------------------------------------------------
# Task 3

print('Task 3: Displacements from frame k_0 to k_0 + 1:')
for i in range(num_points):
    p_0 = trajectories[i, 0]
    q_1 = trajectories[i, 1]
    if np.isnan(q_1):
        print(f"  Point {i}: tracking failed.")
    else:
        d_01 = q_1 - p_0
        print(f"  Point {i} => p={p_0:.2f}, q={q_1:.2f}, displacement d={d_01:.2f}")
```
